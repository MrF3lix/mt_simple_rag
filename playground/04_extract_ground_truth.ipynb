```python
import pandas as pd
```

```python
SOURCE = '../data/kilt_fever_train_small.jsonl'
```

```python
df = pd.read_json(SOURCE, lines=True)
df
```

```python
def extract_wikipedia_link(row):

    item = row['output'][0]['provenance'][0]

    return {
        'wikipedia_id': item['wikipedia_id'],
        'section': item['section'],
        'start_paragraph_id': item['start_paragraph_id'],
        'end_paragraph_id': item['end_paragraph_id']
    }

    
```

```python
df['ground_truth'] = df.apply(extract_wikipedia_link, axis=1)
df[['id', 'input', 'ground_truth']].to_json('truth.json', orient='records')
```
