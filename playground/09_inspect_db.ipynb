```python
import duckdb
```

```python
DB = '../data/kilt_wiki.duckdb'
WIKI_ALL = '../data/wikipedia_kilt.duckdb'
```

```python
con = duckdb.connect(DB)
```

```python
document_id = '23041'
reference_paragraphs = [2, 3, 5]
excluded = [5, 4]

res = con.execute(f"""
    SELECT *
    FROM paragraph
    WHERE wikipedia_id = '{document_id}' AND index IN ({','.join(['?']*len(reference_paragraphs))})
""", reference_paragraphs).df()

res.loc[(~res['index'].isin(excluded) & res['wikipedia_id'] != document_id)].head(5)
```
